## (p.44) Working with a Histogram Representation

```python
image = cv2.imread("./images/panda.jpg")
# plot a histogram

#  : https://blog.naver.com/samsjang/220537529081
# cv2.calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]])
#   image   (uint8 or float32 type). Array.
#   channels   (X ).  graysacle [0], color  B, G, R     [0], [1], [2]  
#   mask   . None  .
#   histSize  BINS . [256]
#   ranges  Range. [0,256]
histogram_image = cv2.calcHist([image], [0], None, [256], [0,256])

# flaten the histogram
plt.hist(histogram_image.ravel(), 256, [0, 256])
plt.show()
```

```python
# view color channels
color = ['b', 'g', 'r']

# seperate the colors and plot the histogram
for i, col in enumerate(color):
    hist = cv2.calcHist([image], [i], None, [256], [0,256])
    plt.plot(hist, color = col)
    plt.xlim([0, 256])
    
plt.show()
```

# (P.46) Videos
## Loading Videos from a Webcam

```python
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    cv2.imshow("frame", frame)
    key = cv2.waitKey(1)
    # 27  Esc 
    if key == 27:
        break    

cap.release()
cv2.destroyAllWindows()
```
